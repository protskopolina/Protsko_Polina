## **Задача 2**

**Задача:** З ймовірністю влучення при одному пострілі $p = 0.7$ стрілок стріляє у мішень до першого влучення, але може виконати не більше 4 пострілів. ДВВ $X$ – кількість промахів. Необхідно: 1) знайти закон розподілу ДВВ; 2) виразити функцію розподілу та функцію щільності розподілу ДВВ за допомогою функції Хевісайда та $\delta$-функції Дірака; 3) побудувати графіки функцій розподілу та щільності розподілу; 4) знайти асиметрію та ексцес.

**Розв'язок:**

$X$ - кількість промахів. Ймовірність влучення $p=0.7$, ймовірність промаху $q=1-p=0.3$.
Можливі значення $X$: $\{0, 1, 2, 3\}$.

**1) Закон розподілу ДВВ**

*   $X=0$ (0 промахів): Перший постріл - влучення.
    $P(X=0) = p = 0.7$
*   $X=1$ (1 промах): Перший - промах, другий - влучення.
    $P(X=1) = q \cdot p = 0.3 \cdot 0.7 = 0.21$
*   $X=2$ (2 промахи): Перші два - промахи, третій - влучення.
    $P(X=2) = q^2 \cdot p = (0.3)^2 \cdot 0.7 = 0.09 \cdot 0.7 = 0.063$
*   $X=3$ (3 промахи): Ця подія відбувається, якщо було 3 промахи і 4-й постріл - влучення, АБО якщо всі 4 постріли - промахи (бо більше стріляти не можна).
    $P(X=3) = q^3 \cdot p + q^4 = q^3(p+q) = q^3 = (0.3)^3 = 0.027$

Таблиця закону розподілу:
| $k$ | 0 | 1 | 2 | 3 |
| :--- | :--- | :--- | :--- | :--- |
| $P(X=k)$ | 0.7 | 0.21 | 0.063 | 0.027 |

*(Перевірка: $0.7 + 0.21 + 0.063 + 0.027 = 1.0$)*

**2) Функції розподілу та щільності**

*   $f(x) = 0.7\delta(x-0) + 0.21\delta(x-1) + 0.063\delta(x-2) + 0.027\delta(x-3)$
*   $F(x) = 0.7H(x-0) + 0.21H(x-1) + 0.063H(x-2) + 0.027H(x-3)$

**3) Графіки функцій**

*   **Графік функції щільності:** Стовпчики в точках 0, 1, 2, 3 з висотами 0.7, 0.21, 0.063, 0.027.

![Графік функції щільності](Photo/5_2_1.png)

*   **Графік функції розподілу:** Східчаста функція, що зростає в точках 0, 1, 2, 3 на величину відповідних ймовірностей.

![Графік функції розподілу](Photo/5_2_2.png)

**4) Асиметрія та ексцес**

Спочатку знайдемо моменти за визначенням:
*   $M(X) = \sum k \cdot P(X=k) = 0(0.7) + 1(0.21) + 2(0.063) + 3(0.027) = 0 + 0.21 + 0.126 + 0.081 = 0.417$
*   $M(X^2) = \sum k^2 \cdot P(X=k) = 0^2(0.7) + 1^2(0.21) + 2^2(0.063) + 3^2(0.027) = 0 + 0.21 + 4(0.063) + 9(0.027) = 0.21 + 0.252 + 0.243 = 0.705$
*   $D(X) = M(X^2) - [M(X)]^2 = 0.705 - (0.417)^2 = 0.705 - 0.173889 = 0.531111$
*   $\sigma(X) = \sqrt{D(X)} \approx \sqrt{0.531111} \approx 0.7288$

Для асиметрії та ексцесу потрібні $\mu_3$ та $\mu_4$:
*   $\mu_3 = \sum (k-M(X))^3 P(k) = (0-0.417)^3(0.7) + (1-0.417)^3(0.21) + (2-0.417)^3(0.063) + (3-0.417)^3(0.027) \approx 0.708$
*   $\mu_4 = \sum (k-M(X))^4 P(k) = (0-0.417)^4(0.7) + (1-0.417)^4(0.21) + (2-0.417)^4(0.063) + (3-0.417)^4(0.027) \approx 1.488$

*   **Асиметрія:** $As = \frac{\mu_3}{\sigma^3} \approx \frac{0.708}{(0.7288)^3} \approx 1.828$
*   **Ексцес:** $Ex = \frac{\mu_4}{\sigma^4} - 3 \approx \frac{1.488}{(0.7288)^4} - 3 \approx 5.28 - 3 = 2.28$

## **Задача 5**

**Задача:** Завод відправив на базу 500 цілих деталей. Ймовірність зіпсування кожної деталі в дорозі $p = 0.002$. Знайти закон розподілу ДВВ $X$, що дорівнює кількості зіпсованих деталей, та знайти ймовірності подій: пошкоджено менше 3 деталей; пошкоджено більше 2 деталей; пошкоджено хоча б одну деталь.

**Розв'язок:**

Маємо велику кількість випробувань $n=500$ і малу ймовірність події $p=0.002$. Це випадок для апроксимації біноміального розподілу **розподілом Пуассона**.
Параметр розподілу $\lambda = n \cdot p = 500 \cdot 0.002 = 1$.
Отже, $X \sim Po(\lambda=1)$.
Формула Пуассона: $P(X=k) = \frac{\lambda^k e^{-\lambda}}{k!}$. Використовуємо $e^{-1} \approx 0.3679$.

**Закон розподілу ДВВ $X$**

*   $P(X=0) = \frac{1^0 e^{-1}}{0!} = e^{-1} \approx 0.3679$
*   $P(X=1) = \frac{1^1 e^{-1}}{1!} = e^{-1} \approx 0.3679$
*   $P(X=2) = \frac{1^2 e^{-1}}{2!} = \frac{e^{-1}}{2} \approx 0.1839$
*   $P(X=3) = \frac{1^3 e^{-1}}{3!} = \frac{e^{-1}}{6} \approx 0.0613$
*   $P(X=4) = \frac{1^4 e^{-1}}{4!} = \frac{e^{-1}}{24} \approx 0.0153$
*   ... і так далі.

**Ймовірності подій**

*   **Пошкоджено менше 3 деталей ($X < 3$):**
    $P(X < 3) = P(X=0) + P(X=1) + P(X=2) \approx 0.3679 + 0.3679 + 0.1839 = 0.9197$
*   **Пошкоджено більше 2 деталей ($X > 2$):**
    $P(X > 2) = 1 - P(X \le 2) = 1 - (P(X=0) + P(X=1) + P(X=2)) \approx 1 - 0.9197 = 0.0803$
*   **Пошкоджено хоча б одну деталь ($X \ge 1$):**
    $P(X \ge 1) = 1 - P(X=0) \approx 1 - 0.3679 = 0.6321$

*   Кількість студентів, що сьогодні відсутні на парі.
*   Кількість бракованих деталей у партії.

**2. Навести декілька прикладів неперервної випадкової величини.**
*   Температура повітря в кімнаті.
*   Час, за який студент добирається до університету.
*   Точна вага людини.
*   Напруга в електромережі.

---

**3. Чи для всіх розподілів існують математичне сподівання і дисперсія?**
Ні, не для всіх. Класичний приклад — **розподіл Коші**, для якого не існує ні математичного сподівання, ні дисперсії, оскільки відповідні інтеграли розбігаються.

---

**4. Як виправдати використання математичного сподівання як числової характеристики для розподілу, який не має скінченного математичного сподівання?**
Це неможливо. Якщо математичне сподівання не існує (є нескінченним або невизначеним), його не можна використовувати як числову характеристику. В таких випадках для опису "центра" розподілу використовують інші характеристики, наприклад, **медіану** або **моду**.

---

**5. Яка форма закону розподілу є універсальною і може бути застосовна як для ДВВ, так і для НВВ?**
**Функція розподілу** $F(x) = P(X < x)$. Вона існує для будь-якої випадкової величини, як дискретної (де вона є східчастою), так і неперервної (де вона є неперервною функцією).

---

**6. Які альтернативні числові характеристики можна використовувати для опису розподілу, якщо математичне сподівання не відображає його повністю?**
*   **Медіана:** значення, яке ділить розподіл навпіл (50% значень менше, 50% — більше).
*   **Мода:** значення, що зустрічається найчастіше.
*   **Квантилі, децилі, перцентилі:** значення, що ділять розподіл на рівні частини (4, 10, 100 відповідно).

---

**7. У чому полягає ймовірнісний та статистичний сенс математичного сподівання?**
*   **Ймовірнісний сенс:** Це середньозважене всіх можливих значень випадкової величини, де вагами є їхні ймовірності.
*   **Статистичний сенс:** Це середнє арифметичне значення, до якого прагне результат при великій кількості повторних експериментів (згідно із законом великих чисел).

---

**8. Чому важливо враховувати асиметрію та ексцес при аналізі розподілу величин?**
*   **Асиметрія (As)** показує "скошеність" розподілу. Додатна асиметрія означає "довгий хвіст" праворуч, від'ємна — ліворуч. Це допомагає зрозуміти, чи переважають екстремально великі чи малі значення.
*   **Ексцес (Ex)** показує "гостроту" піку розподілу порівняно з нормальним. Додатний ексцес ("гострий пік") вказує на більшу ймовірність екстремальних значень ("важкі хвости").

---

**9. Чому, якщо для певної ВВ не існує математичного сподівання, то не існує дисперсія, асиметрія і ексцес? Відповідь обґрунтуйте.**
Тому що всі ці характеристики **обчислюються через математичне сподівання**.
*   **Дисперсія:** $D(X) = M[(X - M(X))^2]$. Якщо не існує $M(X)$, цю формулу неможливо обчислити.
*   **Асиметрія та ексцес:** обчислюються через центральні моменти ($\mu_3, \mu_4$), які, в свою чергу, є математичними сподіваннями степенів відхилення від $M(X)$. Наприклад, $\mu_3 = M[(X - M(X))^3]$. Без $M(X)$ центральні моменти не визначені.

---

**10. Чому на практиці часто можна апріорі вважати розподіл ВВ нормальним?**
Через **Центральну граничну теорему (ЦГТ)**. Вона стверджує, що сума великої кількості незалежних, однаково розподілених випадкових величин (з будь-яким початковим розподілом) буде мати розподіл, близький до нормального. Оскільки багато реальних процесів (наприклад, похибки вимірювань) є результатом дії багатьох незалежних факторів, їхній сумарний ефект часто підкоряється нормальному закону.
