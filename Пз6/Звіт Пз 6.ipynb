## Задача 1

**Задача:**
$X \sim U(a;b)$. Знайти закон розподілу ВВ $Z = \min(X)$.

**Формулювання:** $X \sim U(a;b)$. Знайти закон розподілу ВВ $Z = \min(X_1, \dots, X_n)$, де $X_i$ — незалежна вибірка з розподілу $X$.

**Розв'язок:**

Функція розподілу для мінімуму з $n$ незалежних однаково розподілених випадкових величин знаходиться за формулою:
$$ F_Z(z) = 1 - (1 - F_X(z))^n $$

Для рівномірного розподілу $U(a,b)$ функція розподілу має вигляд:
$$ F_X(z) = \begin{cases} 0, & z < a \\ \frac{z-a}{b-a}, & a \le z \le b \\ 1, & z > b \end{cases} $$

Підставляємо $F_X(z)$ у формулу для $F_Z(z)$ для інтервалу $[a, b]$:
$$ F_Z(z) = 1 - \left(1 - \frac{z-a}{b-a}\right)^n = 1 - \left(\frac{b-a - (z-a)}{b-a}\right)^n = 1 - \left(\frac{b-z}{b-a}\right)^n $$

**Відповідь:**
Функція розподілу для $Z = \min(X_1, \dots, X_n)$:
$$ F_Z(z) = \begin{cases} 0, & z < a \\ 1 - \left(\frac{b-z}{b-a}\right)^n, & a \le z \le b \\ 1, & z > b \end{cases} $$


## Задача 18

**Задача:**
Шляхом експериментальних досліджень встановлено, що похибка, зумовлена шумом аналогових елементів аналого-цифрового перетворювача (АЦП) комп’ютеризованої системи контролю (КСК), – випадкова величина $X$, що має нормальний закон розподілу з параметрами $\mu = 0, \sigma = 1$, а похибка квантування – випадкова величина $Y$, що має рівномірний розподіл з параметрами $a = 0, b = 1$. Знайти закон розподілу сумарної похибки $Z = X + Y$.

**Формулювання:** $X \sim N(0, 1)$, $Y \sim U(0, 1)$. Знайти закон розподілу сумарної похибки $Z = X + Y$.

**Розв'язок:**

Щільність розподілу суми двох незалежних випадкових величин знаходиться через інтеграл згортки:
$$ f_Z(z) = \int_{-\infty}^{\infty} f_X(x) \cdot f_Y(z - x) \,dx $$

Щільності для заданих розподілів:
*   $f_X(x) = \frac{1}{\sqrt{2\pi}} e^{-x^2/2}$
*   $f_Y(y) = \begin{cases} 1, & y \in [0, 1] \\ 0, & \text{в інших випадках} \end{cases}$

Підінтегральний вираз не дорівнює нулю тільки тоді, коли $0 \le z - x \le 1$, що еквівалентно $z - 1 \le x \le z$. Це визначає нові межі інтегрування.

$$ f_Z(z) = \int_{z-1}^{z} \frac{1}{\sqrt{2\pi}} e^{-x^2/2} \cdot 1 \,dx $$

Цей інтеграл виражається через функцію розподілу стандартного нормального закону $\Phi(t) = \int_{-\infty}^{t} \frac{1}{\sqrt{2\pi}} e^{-u^2/2} \,du$.

**Відповідь:**
Щільність розподілу сумарної похибки $Z$:
$$ f_Z(z) = \Phi(z) - \Phi(z-1) $$


## Задача 2

**Задача:**
Випадкові величини $X$ та $Y$ незалежні та обидві мають рівномірний закон розподілу з параметрами $a = 0, b = 2$. Знайти функції розподілу та щільності розподілу випадкової величини $Z = X + Y$.

**Формулювання:** $X, Y \sim U(0, 2)$ незалежні. Знайти функції розподілу та щільності розподілу $Z = X + Y$.

**Розв'язок:**

Використовуємо згортку. Щільності $f_X(x) = f_Y(y) = 1/2$ на інтервалі $[0, 2]$.
$$ f_Z(z) = \int_{-\infty}^{\infty} f_X(x) f_Y(z-x) \,dx = \int \frac{1}{2} \cdot \frac{1}{2} \,dx = \frac{1}{4} \int \,dx $$
Межі інтегрування визначаються перетином умов $0 \le x \le 2$ та $0 \le z-x \le 2$ (тобто $z-2 \le x \le z$).

1.  **Для $0 \le z \le 2$**: межі інтегрування по $x$ від $0$ до $z$.
    $f_Z(z) = \int_{0}^{z} \frac{1}{4} \,dx = \frac{z}{4}$.
2.  **Для $2 < z \le 4$**: межі інтегрування по $x$ від $z-2$ до $2$.
    $f_Z(z) = \int_{z-2}^{2} \frac{1}{4} \,dx = \frac{1}{4}(2 - (z-2)) = \frac{4-z}{4}$.

**Відповідь (щільність):**
$$ f_Z(z) = \begin{cases} z/4, & 0 \le z \le 2 \\ (4-z)/4, & 2 < z \le 4 \\ 0, & \text{в інших випадках} \end{cases} $$

Інтегруючи щільність, знаходимо функцію розподілу $F_Z(z)$:

**Відповідь (функція розподілу):**
$$ F_Z(z) = \begin{cases} 0, & z < 0 \\ z^2/8, & 0 \le z \le 2 \\ 1 - (4-z)^2/8, & 2 < z \le 4 \\ 1, & z > 4 \end{cases} $$


## Задача 3

**Задача:**
Час між запитами до серверу комп’ютерної мережі є випадковою величиною $X$, що має експоненціальний закон розподілу з параметром $\lambda = 10$. З метою дослідження степені використання серверу необхідно встановити закон розподілу максимумів випадкової величини $X$, тобто деякої випадкової величини $Z = \max(X)$.

**Формулювання:** $X \sim Exp(10)$. Встановити закон розподілу $Z = \max(X_1, \dots, X_n)$.

**Розв'язок:**

Функція розподілу для максимуму з $n$ незалежних однаково розподілених випадкових величин знаходиться за формулою:
$$ F_Z(z) = [F_X(z)]^n $$

Для експоненціального розподілу $Exp(\lambda)$ з $\lambda=10$ функція розподілу має вигляд:
$$ F_X(z) = \begin{cases} 1 - e^{-10z}, & z \ge 0 \\ 0, & z < 0 \end{cases} $$

Підставляємо $F_X(z)$ у формулу для $F_Z(z)$.

**Відповідь:**
Функція розподілу для $Z = \max(X_1, \dots, X_n)$:
$$ F_Z(z) = \begin{cases} (1 - e^{-10z})^n, & z \ge 0 \\ 0, & z < 0 \end{cases} $$
де $n$ — кількість спостережень (розмір вибірки).


## Задача 4

**Задача:** 
Випадкова величина $X$ має рівномірний розподіл з параметрами $a = 0, b = \pi$. Знайти функції розподілу та щільності розподілу випадкової величини $Z = \sin(X)$, обчислити математичне сподівання $M(Z)$ та дисперсію $D(Z)$.

**Формулювання:** $X \sim U(0, \pi)$. Знайти $F_Z(z)$, $f_Z(z)$, $M(Z)$, $D(Z)$ для $Z = \sin(X)$.

**Розв'язок:**

1.  **Функція та щільність розподілу:**
    Область значень $Z$ - відрізок $[0, 1]$. Для $z \in [0, 1]$:
    $F_Z(z) = P(Z < z) = P(\sin(X) < z)$.
    На інтервалі $X \in [0, \pi]$ нерівність виконується для $X \in [0, \arcsin(z)) \cup (\pi - \arcsin(z), \pi]$.
    Оскільки $X$ розподілена рівномірно з щільністю $f_X(x) = 1/\pi$, ймовірність потрапляння в інтервал дорівнює його довжині, поділеній на $\pi$.
    $F_Z(z) = \frac{\arcsin(z)}{\pi} + \frac{\pi - (\pi - \arcsin(z))}{\pi} = \frac{2}{\pi}\arcsin(z)$.
    Щільність $f_Z(z) = F'_Z(z) = \frac{d}{dz} \left( \frac{2}{\pi} \arcsin(z) \right) = \frac{2}{\pi\sqrt{1-z^2}}$.

2.  **Математичне сподівання $M(Z)$:**
    $M(Z) = M(\sin(X)) = \int_{0}^{\pi} \sin(x) \cdot f_X(x) \,dx = \int_{0}^{\pi} \sin(x) \cdot \frac{1}{\pi} \,dx$.
    $M(Z) = \frac{1}{\pi} [-\cos(x)]_0^\pi = \frac{1}{\pi} (1 - (-1)) = \frac{2}{\pi}$.

3.  **Дисперсія $D(Z) = M(Z^2) - [M(Z)]^2$:**
    $M(Z^2) = M(\sin^2(X)) = \int_{0}^{\pi} \sin^2(x) \cdot \frac{1}{\pi} \,dx$.
    Використовуємо $\sin^2(x) = \frac{1 - \cos(2x)}{2}$:
    $M(Z^2) = \frac{1}{\pi} \int_{0}^{\pi} \frac{1 - \cos(2x)}{2} \,dx = \frac{1}{2\pi} \left[ x - \frac{\sin(2x)}{2} \right]_0^\pi = \frac{1}{2\pi} (\pi) = \frac{1}{2}$.
    $D(Z) = M(Z^2) - [M(Z)]^2 = \frac{1}{2} - \left(\frac{2}{\pi}\right)^2 = \frac{1}{2} - \frac{4}{\pi^2}$.

**Відповідь:**
*   $F_Z(z) = \frac{2}{\pi} \arcsin(z)$ для $z \in [0, 1]$.
*   $f_Z(z) = \frac{2}{\pi\sqrt{1-z^2}}$ для $z \in [0, 1)$.
*   $M(Z) = \frac{2}{\pi}$.
*   $D(Z) = \frac{1}{2} - \frac{4}{\pi^2}$.

**1. Як знання закону розподілу значень пікових навантажень у комп’ютерній мережі підприємства може допомогти при моделюванні та аналізі пікових навантажень?**

Знання закону розподілу пікових навантажень дозволяє:
*   **Оптимально підбирати обладнання:** Розрахувати необхідну пропускну здатність для обробки 99.9% піків, уникаючи зайвих витрат або недостатньої потужності.
*   **Реалістично тестувати мережу:** Створювати правдоподібні сценарії стрес-тестів для перевірки відмовостійкості.
*   **Ефективно керувати трафіком (QoS):** Налаштовувати пріоритети для критичних сервісів під час очікуваних перевантажень.
*   **Прогнозувати модернізацію:** Планувати розширення мережі, аналізуючи, як змінюється розподіл навантажень з часом.

---

**2. Як знайти математичне сподівання функції одного випадкового аргумента?**

Математичне сподівання функції $Z = g(X)$ знаходиться за "теоремою про середнє", без необхідності обчислювати закон розподілу $Z$.

*   **Для неперервної ВВ:** $M(g(X)) = \int_{-\infty}^{\infty} g(x) \cdot f_X(x) \,dx$
*   **Для дискретної ВВ:** $M(g(X)) = \sum_{i} g(x_i) \cdot p_i$

---

**3. Як знайти дисперсію функції одного випадкового аргумента?**

Дисперсія функції $Z = g(X)$ знаходиться за стандартною формулою $D(Z) = M(Z^2) - [M(Z)]^2$. Для цього потрібно:
1.  Обчислити $M(Z) = M(g(X))$.
2.  Обчислити $M(Z^2) = M(g^2(X))$.
3.  Підставити знайдені значення у формулу.

---

**4. Чому на етапі обчислення закону розподілу функції від випадкової величини виконати аналіз монотонності функції?**

Аналіз монотонності визначає складність задачі.
*   **Якщо функція монотонна,** рівняння $g(x) = z$ має **один** розв'язок $x = g^{-1}(z)$, що спрощує розрахунки.
*   **Якщо функція немонотонна,** рівняння $g(x) = z$ має **декілька** розв'язків. При обчисленні ймовірності $P(Z<z)$ потрібно враховувати всі інтервали значень $X$, що задовольняють нерівність, і це ускладнює розв'язання.
