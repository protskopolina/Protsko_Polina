## **Задача 18**

**Задача:** Задано матрицю переходу $P_1 = \begin{pmatrix} 0.6 & 0.4 \\ 0.1 & 0.9 \end{pmatrix}$. Знайти матрицю переходу $P_3$.

**Розв'язок:**

Матриця переходу $P_n$ за $n$ кроків знаходиться шляхом піднесення матриці переходу за один крок $P_1$ до степеня $n$. Отже, нам потрібно знайти $P_3 = (P_1)^3$.

Ми можемо зробити це послідовно: спочатку знайдемо $P_2 = (P_1)^2$, а потім $P_3 = P_2 \cdot P_1$.

**Крок 1: Обчислення матриці переходу за 2 кроки ($P_2$)**

$P_2 = (P_1)^2 = P_1 \cdot P_1 = \begin{pmatrix} 0.6 & 0.4 \\ 0.1 & 0.9 \end{pmatrix} \cdot \begin{pmatrix} 0.6 & 0.4 \\ 0.1 & 0.9 \end{pmatrix}$

Обчислюємо кожен елемент матриці $P_2$:
*   $p_{11}^{(2)} = (0.6 \cdot 0.6) + (0.4 \cdot 0.1) = 0.36 + 0.04 = 0.40$
*   $p_{12}^{(2)} = (0.6 \cdot 0.4) + (0.4 \cdot 0.9) = 0.24 + 0.36 = 0.60$
*   $p_{21}^{(2)} = (0.1 \cdot 0.6) + (0.9 \cdot 0.1) = 0.06 + 0.09 = 0.15$
*   $p_{22}^{(2)} = (0.1 \cdot 0.4) + (0.9 \cdot 0.9) = 0.04 + 0.81 = 0.85$

Таким чином, матриця переходу за 2 кроки:
$P_2 = \begin{pmatrix} 0.40 & 0.60 \\ 0.15 & 0.85 \end{pmatrix}$

*(Перевірка: сума елементів у кожному рядку дорівнює 1. $0.40+0.60=1.0$; $0.15+0.85=1.0$. Все правильно.)*

**Крок 2: Обчислення матриці переходу за 3 кроки ($P_3$)**

$P_3 = P_2 \cdot P_1 = \begin{pmatrix} 0.40 & 0.60 \\ 0.15 & 0.85 \end{pmatrix} \cdot \begin{pmatrix} 0.6 & 0.4 \\ 0.1 & 0.9 \end{pmatrix}$

Обчислюємо кожен елемент матриці $P_3$:
*   $p_{11}^{(3)} = (0.40 \cdot 0.6) + (0.60 \cdot 0.1) = 0.24 + 0.06 = 0.30$
*   $p_{12}^{(3)} = (0.40 \cdot 0.4) + (0.60 \cdot 0.9) = 0.16 + 0.54 = 0.70$
*   $p_{21}^{(3)} = (0.15 \cdot 0.6) + (0.85 \cdot 0.1) = 0.09 + 0.085 = 0.175$
*   $p_{22}^{(3)} = (0.15 \cdot 0.4) + (0.85 \cdot 0.9) = 0.06 + 0.765 = 0.825$

Таким чином, матриця переходу за 3 кроки:
$P_3 = \begin{pmatrix} 0.30 & 0.70 \\ 0.175 & 0.825 \end{pmatrix}$

*(Перевірка: $0.30+0.70=1.0$; $0.175+0.825=1.0$. Все правильно.)*

**Відповідь:** $P_3 = \begin{pmatrix} 0.30 & 0.70 \\ 0.175 & 0.825 \end{pmatrix}$.


## **Задача 19**

**Задача:** Побудувати граф станів СМО «n-клієнтів – Web-сервер» (система М/М/1) і систему рівнянь Колмогорова для $n = 3$, $\lambda = 2$, $\mu = 2$. Знайти $P_{зан}$, $P_0$, $A$, $w$, $T_{обс}$, $T_{відг}$.

**Розв'язок:**

Маємо систему масового обслуговування (СМО) типу M/M/1 з обмеженою кількістю клієнтів (джерел заявок) $n=3$.
*   $\lambda = 2$ — інтенсивність надходження заявок від **одного** вільного клієнта.
*   $\mu = 2$ — інтенсивність обслуговування заявок сервером.

**Крок 1: Побудова графа станів**

Стан системи $S_k$ визначається кількістю клієнтів у системі (на обслуговуванні або в черзі). Оскільки всього клієнтів 3, можливі стани:
*   $S_0$: система вільна (0 клієнтів у системі).
*   $S_1$: 1 клієнт обслуговується, 0 в черзі.
*   $S_2$: 1 клієнт обслуговується, 1 в черзі.
*   $S_3$: 1 клієнт обслуговується, 2 в черзі.

Інтенсивності переходів:
*   Інтенсивність надходження заявок $\lambda_k$ залежить від кількості вільних клієнтів. Якщо в системі $k$ клієнтів, то вільних — $(n-k)$. Отже, $\lambda_k = (n-k)\lambda$.
    *   $\lambda_0 = (3-0)\lambda = 3\lambda = 3 \cdot 2 = 6$
    *   $\lambda_1 = (3-1)\lambda = 2\lambda = 2 \cdot 2 = 4$
    *   $\lambda_2 = (3-2)\lambda = 1\lambda = 1 \cdot 2 = 2$
    *   $\lambda_3 = (3-3)\lambda = 0$ (коли всі клієнти в системі, нові заявки не надходять).
*   Інтенсивність обслуговування $\mu_k$ не залежить від стану, оскільки сервер один (M/M/1).
    *   $\mu_1 = \mu = 2$
    *   $\mu_2 = \mu = 2$
    *   $\mu_3 = \mu = 2$

**Крок 2: Система рівнянь Колмогорова**

Для стаціонарного режиму ($P'_k(t) = 0$) рівняння мають вигляд "потік ймовірності, що входить у стан = потік ймовірності, що виходить зі стану".
*   Для $S_0$: $\mu_1 P_1 = \lambda_0 P_0 \implies 2 P_1 = 6 P_0$
*   Для $S_1$: $\lambda_0 P_0 + \mu_2 P_2 = (\lambda_1 + \mu_1) P_1 \implies 6 P_0 + 2 P_2 = (4+2) P_1 \implies 6 P_0 + 2 P_2 = 6 P_1$
*   Для $S_2$: $\lambda_1 P_1 + \mu_3 P_3 = (\lambda_2 + \mu_2) P_2 \implies 4 P_1 + 2 P_3 = (2+2) P_2 \implies 4 P_1 + 2 P_3 = 4 P_2$
*   Для $S_3$: $\lambda_2 P_2 = \mu_3 P_3 \implies 2 P_2 = 2 P_3$

Додаємо умову нормування: $P_0 + P_1 + P_2 + P_3 = 1$.

**Крок 3: Знаходження ймовірностей станів**

Виразимо всі ймовірності через $P_0$:
1.  З першого рівняння: $P_1 = \frac{6}{2} P_0 = 3 P_0$.
2.  З останнього рівняння: $P_3 = \frac{2}{2} P_2 = P_2$.
3.  Підставимо $P_3=P_2$ у третє рівняння: $4 P_1 + 2 P_2 = 4 P_2 \implies 4 P_1 = 2 P_2 \implies P_2 = 2 P_1$.
4.  Тепер підставимо вираз для $P_1$: $P_2 = 2 (3 P_0) = 6 P_0$.
5.  Оскільки $P_3 = P_2$, то $P_3 = 6 P_0$.

Тепер маємо всі ймовірності, виражені через $P_0$:
*   $P_1 = 3 P_0$
*   $P_2 = 6 P_0$
*   $P_3 = 6 P_0$

Підставляємо в умову нормування:
$P_0 + 3 P_0 + 6 P_0 + 6 P_0 = 1$
$16 P_0 = 1 \implies P_0 = \frac{1}{16}$

Тепер знаходимо решту ймовірностей:
*   $P_0 = \frac{1}{16} = 0.0625$
*   $P_1 = 3 \cdot \frac{1}{16} = \frac{3}{16} = 0.1875$
*   $P_2 = 6 \cdot \frac{1}{16} = \frac{6}{16} = \frac{3}{8} = 0.375$
*   $P_3 = 6 \cdot \frac{1}{16} = \frac{6}{16} = \frac{3}{8} = 0.375$

**Крок 4: Розрахунок характеристик СМО**

*   **$P_{зан}$ (ймовірність того, що сервер зайнятий):**
    Це ймовірність того, що система не в стані $S_0$.
    $P_{зан} = 1 - P_0 = 1 - \frac{1}{16} = \frac{15}{16} = 0.9375$

*   **$P_0$ (ймовірність того, що сервер вільний):**
    $P_0 = \frac{1}{16} = 0.0625$

*   **$A$ (абсолютна пропускна здатність):**
    Середня кількість заявок, що обслуговуються за одиницю часу.
    $A = \mu \cdot P_{зан} = 2 \cdot \frac{15}{16} = \frac{30}{16} = \frac{15}{8} = 1.875$

*   **$w$ (середня кількість клієнтів у черзі):**
    $w = \sum_{k=m+1}^{n} (k-m) P_k$, де $m=1$ (кількість серверів).
    $w = (2-1)P_2 + (3-1)P_3 = 1 \cdot P_2 + 2 \cdot P_3 = \frac{6}{16} + 2 \cdot \frac{6}{16} = \frac{6+12}{16} = \frac{18}{16} = \frac{9}{8} = 1.125$

*   **$T_{обс}$ (середній час обслуговування):**
    $T_{обс} = \frac{1}{\mu} = \frac{1}{2} = 0.5$

*   **$T_{відг}$ (середній час відповіді / перебування клієнта в системі):**
    За формулою Літтла для замкнених систем: $A \cdot T_{відг} = n - w$.
    $T_{відг} = \frac{n - w}{A} = \frac{3 - 1.125}{1.875} = \frac{1.875}{1.875} = 1$

**Відповіді:**
*   $P_{зан} = \frac{15}{16} = 0.9375$
*   $P_0 = \frac{1}{16} = 0.0625$
*   $A = 1.875$
*   $w = 1.125$
*   $T_{обс} = 0.5$
*   $T_{відг} = 1$

# Контрольні питання

**1. Що таке СМО і які основні елементи входять у її структуру?**
**СМО (Система масового обслуговування)** — це система, що аналізує процеси, де виникають черги. Основні елементи: джерело заявок, потік заявок, черга (накопичувач), канали обслуговування та потік обслужених заявок.

**2. Які властивості має найпростіший потік подій, і які його характеристики можна виміряти?**
Найпростіший (Пуассонівський) потік має три властивості: **стаціонарність** (інтенсивність не залежить від часу), **відсутність післядії** (майбутнє не залежить від минулого) та **ординарність** (ймовірність двох подій в малий проміжок часу близька до нуля). Головна характеристика — **інтенсивність потоку (λ)**, середня кількість подій за одиницю часу.

**3. Які основні характеристики СМО визначають її продуктивність?**
*   **Абсолютна пропускна здатність (A):** середня кількість обслужених заявок за одиницю часу.
*   **Середній час перебування в системі (T_відг):** час від прибуття до повного обслуговування.
*   **Середня кількість заявок у черзі (w)** та в системі.
*   **Ймовірність відмови (P_відм):** для систем з обмеженою чергою.

**4. Які фактори впливають на інтенсивність потоку подій в системі масового обслуговування?**
Кількість джерел заявок, їхній стан (активні/неактивні), час доби, день тижня, сезонність, рекламні акції, технічні збої — будь-які зовнішні та внутрішні фактори, що змінюють частоту надходження запитів.

**5. Як визначається інтенсивність обслуговування в СМО?**
**Інтенсивність обслуговування (μ)** — це середня кількість заявок, яку один канал може обслужити за одиницю часу. Це величина, обернена до середнього часу обслуговування однієї заявки ($μ = 1 / T_{обс}$).

**6. Які властивості мають ланцюги Маркова, і як вони застосовуються в теорії СМО?**
Головна властивість — **відсутність післядії (Марковська властивість):** майбутній стан системи залежить тільки від поточного стану, а не від попередньої історії. В СМО вони застосовуються для моделювання переходів системи між станами (наприклад, $S_0 \to S_1 \to S_2$), що дозволяє розрахувати ймовірності цих станів.

**7. Що таке стаціонарний режим роботи СМО і чому він важливий для аналізу?**
**Стаціонарний режим** — це стан, коли ймовірності перебування системи в кожному зі станів ($P_0, P_1, ...$) не змінюються з часом. Він важливий, бо дозволяє розрахувати усереднені довгострокові характеристики продуктивності СМО (пропускну здатність, середній час очікування тощо).

**8. Як визначається ймовірність втрати заявки в системі масового обслуговування?**
Це **ймовірність відмови (P_відм)**, яка дорівнює ймовірності того, що заявка надійде в момент, коли система повністю заповнена (всі канали зайняті і черга заповнена). Для системи з максимальною кількістю місць $N$ це $P_{відм} = P_N$.

**9. Що таке ефективність обслуговування в СМО і як її вимірюють?**
Ефективність — це показник того, наскільки добре система справляється зі своїм завданням. Вимірюється через різні характеристики: мінімізація часу очікування, максимізація пропускної здатності, низька ймовірність відмови, високий коефіцієнт завантаження каналів.

**10. Як визначається коефіцієнт завантаження системи масового обслуговування, і чому він важливий для оцінки її продуктивності?**
**Коефіцієнт завантаження (ρ)** — це відношення інтенсивності потоку заявок до сумарної інтенсивності обслуговування ($\rho = \lambda / (m \cdot \mu)$, де $m$ — кількість каналів). Він показує, яка частина ресурсу системи використовується. Якщо $\rho \ge 1$ (для систем з нескінченною чергою), черга буде нескінченно зростати. Це ключовий показник стабільності системи.
